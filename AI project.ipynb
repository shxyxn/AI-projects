```python
#Shayan Eskandari        99573103

import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# dataset1 :
x1 = np.random.rand(5,1) *10 -5
y1 = 2 * x1 + 5
y1 = y1 + np.random.normal(scale=1, size=y1.shape)
plt.plot(x1, y1, 'bo')
```

```python
# DEGREE = 1

poly_features = PolynomialFeatures(degree=1, include_bias=False)
x1_poly = poly_features.fit_transform(x1)
reg1_1 = LinearRegression()
reg1_1.fit(x1_poly, y1)
```

```python
# x1_vals = np.linspace(-5,5,5).reshape(-1,1)
x1_vals = x1
x1_vals_poly = poly_features.transform(x1_vals)
y1_vals = reg1_1.predict(x1_vals_poly)
plt.suptitle('DATA 1 : Degree 1')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(x1, y1)
plt.plot(x1_vals, y1_vals, color="r")
plt.show()
```

```python
len(y1_vals)
```

```python
print(f" intercept : {reg1_1.intercept_} \n coefficient : {reg1_1.coef_} \n MSE : {metrics.mean_squared_error(y1, y1_vals)} \n R2 Score : {r2_score(y1, y1_vals)}")
```

```python
## DEGREE = 4

poly_features_4 = PolynomialFeatures(degree=4, include_bias=False)
x1_poly_4 = poly_features_4.fit_transform(x1)
reg1_1_4 = LinearRegression()
reg1_1_4.fit(x1_poly_4, y1)
```

```python
x1_vals_4_show = np.linspace(-5,5,500).reshape(-1,1)
x1_vals_poly_4_show = poly_features_4.transform(x1_vals_4_show)
y1_vals_4_show = reg1_1_4.predict(x1_vals_poly_4_show)
x1_vals_4 = x1
x1_vals_poly_4 = poly_features_4.transform(x1_vals_4)
y1_vals_4 = reg1_1_4.predict(x1_vals_poly_4)
plt.suptitle('DATA 1 : Degree 4')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(x1, y1)
plt.plot(x1_vals_4_show, y1_vals_4_show, color="r")
plt.show()
```

```python
print(f" intercept : {reg1_1_4.intercept_} \n coefficient : {reg1_1_4.coef_} \n MSE : {metrics.mean_squared_error(y1, y1_vals_4)} \n R2 Score : {r2_score(y1, y1_vals_4)}")
```

```python
# DEGREE = 16

poly_features_16 = PolynomialFeatures(degree=16, include_bias=False)
x1_poly_16 = poly_features_16.fit_transform(x1)
reg1_1_16 = LinearRegression()
reg1_1_16.fit(x1_poly_16, y1)
x1_vals_16_show = np.linspace(-5,5,100).reshape(-1,1)
x1_vals_poly_16_show = poly_features_16.transform(x1_vals_16_show)
y1_vals_16_show = reg1_1_16.predict(x1_vals_poly_16_show)
x1_vals_16 = x1
x1_vals_poly_16 = poly_features_16.transform(x1_vals_16)
y1_vals_16 = reg1_1_16.predict(x1_vals_poly_16)
plt.suptitle('DATA 1 : Degree 16')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(x1, y1)
plt.plot(x1_vals_16_show, y1_vals_16_show, color="r")
plt.show()
```
